### Explorasi Data Analysis (EDA) terkait penjualan produk dan kataegori

### Explorasi Data untuk menghitung Total Nilai Pembayaran per Metode Pembayaran

### Explorasi Data untuk menghitung Rata-Rata Nilai Pembayaran per Metode Pembayaran

```python
# Menghitung rata-rata nilai pembayaran per metode pembayaran
average_payment_value = order_payments_df.groupby('payment_type')['payment_value'].mean().reset_index()
average_payment_value.columns = ['payment_type', 'average_payment_value']

# Menampilkan hasil
print(average_payment_value.head(10))
```

### Explorasi Data untuk menghitung Rata-Rata Waktu Pengiriman Berdasarkan Lokasi Penjual dan Pelanggan

```python
# Gabungkan data
order_data = pd.merge(orders_df, order_items_df, on='order_id', how='inner')
order_data = pd.merge(order_data, sellers_df, on='seller_id', how='inner')
order_data = pd.merge(order_data, customers_df, on='customer_id', how='inner')

# Konversi kolom waktu ke tipe datetime
order_data['order_purchase_timestamp'] = pd.to_datetime(order_data['order_purchase_timestamp'])
order_data['order_approved_at'] = pd.to_datetime(order_data['order_approved_at'])

# Hitung waktu pengiriman dalam jam
order_data['delivery_time'] = (order_data['order_approved_at'] - order_data['order_purchase_timestamp']).dt.total_seconds() / 3600

# Tambahkan lokasi penjual dan pelanggan
order_data['seller_location'] = order_data['seller_city'] + ', ' + order_data['seller_state']
order_data['customer_location'] = order_data['customer_city'] + ', ' + order_data['customer_state']

# Hitung rata-rata waktu pengiriman berdasarkan lokasi penjual dan pelanggan
avg_delivery_time = order_data.groupby(['seller_location', 'customer_location'])['delivery_time'].mean().reset_index()

# Ubah nama kolom
avg_delivery_time.columns = ['seller_location', 'customer_location', 'average_delivery_time']

# Menampilkan hasil
print(avg_delivery_time.head(10))

```

### Explorasi Data dan Analisis Penjualan Geospasial Berdasarkan Lokasi dan Kategori Produk

```python
# 1. Menggabungkan Data
merged_data = (
    orders_df.merge(order_items_df, on='order_id', how='inner')
              .merge(products_df, on='product_id', how='inner')
              .merge(customers_df, on='customer_id', how='inner')
              .merge(geolocation_df, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
)

# Step 1: Determine the top 5 product categories by sales
top_categories = merged_data.groupby('product_category_name')['price'].sum().nlargest(5).index.tolist()
print(top_categories)

# Step 2: Determine the top 5 customer locations by sales
top_locations = merged_data.groupby('customer_city')['price'].sum().nlargest(5).index.tolist()
print(top_locations)

# Step 3: Filter the merged_data for the top 5 categories and locations
filtered_data = merged_data[(merged_data['product_category_name'].isin(top_categories)) &
                            (merged_data['customer_city'].isin(top_locations))]

# Step 4: Group the data by customer location and product category for visualization
sales_by_category_location = filtered_data.groupby(['customer_city', 'product_category_name'])['price'].sum().reset_index()
sales_by_category_location.columns = ['customer_city', 'product_category_name', 'total_sales']

# Step 5: Tampilkan hasil
print(sales_by_category_location.head(20
                                    ))
```
